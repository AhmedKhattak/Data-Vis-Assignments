Processing and finding the missing values through python.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
np.random.seed(69) # set seed
```

Read data into dataframe

```python
df1 = sns.load_dataset("iris")
df2 = sns.load_dataset("car_crashes")
df3 = sns.load_dataset("tips")
```

Show data

```python
print(df1.head()) # iris dataset
print("\n")
print(df2.head()) # car_crashes dataset
print("\n")
print(df3.head()) # restraunt tips dataset
```

Add missing values to one continous variable in each dataset

```python
df1_missing = 40 # the number of missing data points in a particular column/ variable
df2_missing = 40
df3_missing = 40

df1_col = 'petal_width'
df2_col = 'speeding'
df3_col = 'total_bill'

x = np.random.choice(df1.shape[0], df1_missing, replace = False)
df1_na = df1.copy() # do not remove these calls !!
df1_na.loc[x,df1_col] = np.nan

x = np.random.choice(df2.shape[0], df2_missing, replace = False)
df2_na =  df2.copy()
df2_na.loc[x,df2_col] = np.nan

x = np.random.choice(df3.shape[0], df3_missing, replace = False)
df3_na = df3.copy()
df3_na.loc[x,df3_col] = np.nan
```

Impute using mean, median, mode

```python
# df1 mean, median, mode

df1_mean, df1_median, df1_mode = df1_na.copy(), df1_na.copy(), df1_na.copy()
df2_mean, df2_median, df2_mode = df2_na.copy(), df2_na.copy(), df2_na.copy()
df3_mean, df3_median, df3_mode = df3_na.copy(), df3_na.copy(), df3_na.copy()



# df1 mean, median, mode
df1_mean.loc[df1_mean[df1_col].isnull(),df1_col] = df1_mean[df1_col].mean()
df1_median.loc[df1_median[df1_col].isnull(),df1_col]  = df1_median[df1_col].median()
df1_mode.loc[df1_mode[df1_col].isnull(),df1_col]  = df1_mode[df1_col].mode()[0]

#df2 mean , median, mode
df2_mean.loc[df2_mean[df2_col].isnull(),df2_col] = df2_mean[df2_col].mean()
df2_median.loc[df2_median[df2_col].isnull(),df2_col]  = df2_median[df2_col].median()
df2_mode.loc[df2_mode[df2_col].isnull(),df2_col]  = df2_mode[df2_col].mode()[0]

# df3 mean , median, mode
df3_mean.loc[df3_mean[df3_col].isnull(),df3_col] = df3_mean[df3_col].mean()
df3_median.loc[df3_median[df3_col].isnull(),df3_col]  = df3_median[df3_col].median()
df3_mode.loc[df3_mode[df3_col].isnull(),df3_col]  = df3_mode[df3_col].mode()[0]
```

Check columns with missing values

```python
print("0 indicate no missing value in that column\n")

print("IRIS")
print("-" * 25)
print(df1_na.isnull().sum(axis = 0))
print()
print("CAR CRASHES")
print("-" * 25)
print(df2_na.isnull().sum(axis = 0))
print()
print("RESTRAUNT TIPS")
print("-" * 25)
print(df3_na.isnull().sum(axis = 0))
```

Visualize continous variable before and after imputation using histograms

Visualize IRIS

```python
font_size = {'fontsize': 16}
fig = plt.figure(0, figsize = (20,20))
ax1 = plt.subplot2grid((2, 3), (0, 0), colspan = 4)
ax2 = plt.subplot2grid((2, 3), (1, 0))
ax3 = plt.subplot2grid((2, 3), (1, 1))
ax4 = plt.subplot2grid((2, 3), (1, 2))

fig.suptitle('Histograms of petal-widths from IRIS data set before and after imputation via mean, median, mode', fontsize = 16)
sns.distplot(df1[df1_col], kde = False, ax = ax1).set_title('Orignal',  fontdict = font_size) # orginal data
sns.distplot(df1_mean[df1_col], kde = False, ax = ax2).set_title('Mean',  fontdict = font_size) # mean
sns.distplot(df1_median[df1_col], kde = False, ax = ax3).set_title('Median',  fontdict = font_size) #median
sns.distplot(df1_mode[df1_col], kde = False, ax = ax4).set_title('Mode',  fontdict = font_size) #mode
```
