Create a function to add missingness to multiple columns randomly with minimal overlap

Note: Any function where there is a .copy() call on a dataframe may not work properly in pandas pipelines

```python
def nullpute(dataframe: DataFrame, percent_missing = 0.20, columns = None):
    # perform early exit
    if columns is None:
        print("Columns must be given !")
        return
    for col in columns:
        if(is_numeric_dtype(dataframe[col]) != True):
            print(f'At least one column "{col}" in not numeric all columns must be numeric dtype')
            return
    dataframe = dataframe.copy() # need to copy so that this function is non mutating
    # all checks have passed now perform operations
    num_rows = len(dataframe) # num of rows of dataframe
    missing_rows = int(percent_missing * num_rows) # num of rows to remove by percent
    # random_rows = np.random.randint(low = 0, 
    #                                 high = num_rows, 
    #                                 size = missing_rows)
    # Use random choice without replacement
    random_rows_per_column = [np.random.choice(num_rows,missing_rows, 
                                              replace = False) for cols in columns]
    
    
    for x,y in zip(random_rows_per_column, columns):
        dataframe.loc[x,y] = np.nan
    # dataframe.loc[random_rows,columns] = np.nan
    print(f"{missing_rows} observations have been replaced with NaN for columns {columns}")
    return dataframe
  
```

## Impute missing Data

Create a function to streamline imputation

```python
def impute(dataframe: DataFrame, columns = None, method = None):
    # perform early exit
    
    methods = ['mean','median','mode']
    
    if (method is None) or (method in methods == False):
        print(f"User must provide a imputation method which are {methods} ")
    if  columns is None:
        print('User must provide columns to impute on')
        return
    for col in columns:
        if(is_numeric_dtype(dataframe[col]) != True):
            print(f'At least one column "{col}" in not numeric all columns must be numeric dtype')
            return
        
    dataframe = dataframe.copy() # create copy of dataframe
    # impute each colum of dataframe and then return a copy 
    for col in columns:
        
        if method == 'mean':
            imputed_result = dataframe[col].mean()
        elif method == 'median':
            imputed_result = dataframe[col].median()
        elif method == 'mode':
            imputed_result = dataframe[col].mode()[0]
       
        dataframe[dataframe[col].isnull()]  = imputed_result
    return dataframe
   
```

Create a KNN imputation function under the hood it uses sklearn KNeighborsRegressor

```python
from sklearn.neighbors import KNeighborsRegressor

def impute_knn(dataframe: DataFrame, target = None , columns = None, **kwargs):
    # early exit
    if target is None or columns is None:
        print("target and/or columns cannot be null")
        return
    
    # copy dataframe  
    dataframe = dataframe.copy()
    # create training and testing pairs
    columns.insert(0, target)
    testing = dataframe[columns].values
    training = dataframe[columns].dropna(axis=0, how='any').values
    #training = df1[['petal-width','petal-length','sepal-width','sepal-length']].values
    # create classifier
    clf = KNeighborsRegressor(**kwargs)
    # train model
    trained_model = clf.fit(training[:,1:], training[:,0])
    # perform regression
    result = trained_model.predict(testing[:,1:])
    # transform resultant numpy array to series
    res_series = pd.Series(result, name = "result")
    # for each missing value replace it with its corresponding regressed value
    dataframe.loc[dataframe[target].isnull(), target]  = res_series
    return dataframe
    
    
```

Create argument dictionaries for imputation using mean, median, mode and knn regression
